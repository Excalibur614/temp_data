```python
import pandas as pd
pd.set_option('display.max_columns', None)
key_cols = ['LastPrice', 'Volume', 'Turnover', 'BPrice1', 'BVol1', 'SPrice1', 'SVol1', 'BPrice2', 'BVol2', 'SPrice2', 'SVol2', 'BPrice3', 'BVol3', 'SPrice3', 'SVol3', 'SPrice4', 'BVol4', 'SPrice4', 'SVol4', 'BPrice5', 'BVol5', 'SPrice5', 'SVol5']

# IF 沪深300指数期货
df = pd.read_pickle('2023.07.20/CFFEX_IF2307')
df['ReceiveTime'] = pd.to_datetime(df['ReceiveTime'], unit='us').dt.tz_localize(tz='UTC').dt.tz_convert(tz='Asia/Shanghai') 
df = df.set_index('ReceiveTime')[key_cols]

display(df.head())
df.describe()
```
